```python
tableau_pokemons=page.find_all('div',"infocard-list infocard-list-pkmn-lg")
liste_des_pokemons=[]
for gen in tableau_pokemons :
    rows= gen.find_all("div")   
    for row in rows :
        t=row.find("a","ent-name")
        liste_des_pokemons.append(t.text)
print(len(liste_des_pokemons),liste_des_pokemons)
```

```python
def get_stats(pokemon):   
    url=("https://pokemondb.net/pokedex/"+pokemon)

    pokemontext = requests.get(url).content
    soup= bs4.BeautifulSoup(pokemontext, "lxml")
    caracteristics=soup.find_all("table","vitals-table")

    dico={ 'Name' : pokemon}
    for j in range(3):
        cols1=caracteristics[j].find_all("th")
        cols2=caracteristics[j].find_all("td")
        #print(cols2[0].text.strip(),len(cols2),len(cols1))

        for i in range(len(cols1)):
            dico[cols1[i].text.strip()]=cols2[i].text.strip()

    cols1=caracteristics[3].find_all("th")
    cols2=caracteristics[3].find_all("td","cell-num")

    cols2=[cols2[i] for i in range(len(cols2)) if i%3==0 ]
    #print(cols1,len(cols1),len(cols2))
    for i in range(len(cols1)):
        if i<len(cols2):
            dico[cols1[i].text.strip()]=cols2[i].text.strip()


    return(dico)

get_stats("Ivysaur")
```

```python
def extract_keys_and_values(dictionnaire):
    # Récupérer les clés et les valeurs sous forme de listes
    clefs = list(dictionnaire.keys())
    valeurs = list(dictionnaire.values())
    return clefs, valeurs

get_stats("nidoran-f")
```

```python
import re

def filtrer_lettres(chaine):
    return ''.join(re.findall(r'[a-zA-Z-]', chaine))


filtrer_lettres("ho-oh")
```

```python
import pandas as pd
pokedex={'X' : list(get_stats("ivysaur").keys())}
for b in liste_des_pokemons:
    print(b)
    if b[-1]=='♀' :
        b=b[:len(b)-1]+'-f'
        pokedex[b]=list(get_stats(b).values())
    elif b[-1]=='♂':
        b=b[:len(b)-1]+'-m'
    elif b=='Mr. Mime' :
        b='mr-mime' 
        pokedex[b]=list(get_stats(b).values())
    elif b=='Mr. Rime' :
        b='mr-rime' 
        pokedex[b]=list(get_stats(b).values())
    else: pokedex[b]=list(get_stats(filtrer_lettres(b)).values())
data_pokemon = pd.DataFrame.from_dict(pokedex, orient="index")
data_pokemon.head()
```
